## 1. Generar una imagen de una función trigronométrica (función bidimensional) 
### **Sinusoidal Gratings**

<p align="justify">
En teoría, cualquier imagen puede ser reconstruida a partir de ondas senoidales. Es bueno entender como podemos generar una imagen que nos muestra una onda senoidal. 


```python
import numpy as np
import matplotlib.pyplot as plt

# Funcion senoidal
amplitud = 1.0
frecuencia = 2*np.pi / 50   # 50 = longitud de onda
fase = 0

# Crear cuadricula de puntos en plano xy
x = np.linspace(-100, 100, 500)
X, Y = np.meshgrid(x, x)

# Calcular los valores de la función senoidal en cada punto de la cuadrícula
fx = amplitud * np.sin(frecuencia * X + fase)

# Graficar la imagen de la función senoidal
plt.imshow(fx, cmap='gray', extent=[x.min(), x.max(), x.min(), x.max()])    # cmap = muestra la imagen en tonos de grises
plt.colorbar(label='Amplitud')                                              # extente = especifica valores minimos y maximos de los ejes
plt.title('Imagen de funció senoidal bidimensional')
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.show()
```

## 2. Generar señales


### 2. Genera una señal: estas señales las vas a utilizar en los puntos siguientes

Genera la señal $$ sinx+cos2x $$ y graficala.

### 2.2 Genera otra señal no pura como la anterior, sino con ruido gaussiano añadido.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores de x
x = np.linspace(0, 2 * np.pi, 1000)

# Senal senoidal
coseno = np.cos(x**2)

# Parametros para senal no pura
amplitud = 1.0
frecuencia = 1
fase = np.pi/3

# Calcular una señal no pura combinando las funciones senoidales y cosenoidales
gx = amplitud * np.sin(2 * np.pi * frecuencia * x + fase) + coseno

# Graficar la señal no pura
plt.plot(x, gx, color='red')
plt.title(r'Gráfica $ g(x) = A \cdot \sin(2\pi\cdot x + \frac{\pi}{3}) + \sin(x^2)$')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid(True)
plt.show()
```

### 3.1 Obtenga la transformada de Fourier para las señales que generaste en el  punto anterior


```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores de x
x = np.linspace(0, 2 * np.pi, 1000)

# Calcular senal f(x) = sin(x) + cos(2x)
fx = np.sin(x) + np.cos(2 * x)

# Calcular trasnformada de Fourier
Fourier_f = np.fft.fft(fx)

# Obtener frecuencias de la transformada
frec_fourier_f = np.fft.fftfreq(len(x), x[1]- x[0])

# Graficar la transformada de Fourier
plt.plot(frec_fourier_f, np.abs(Fourier_f), color='black')
plt.title('Transformada de Fourier de f(x)')
plt.xlabel('Frecuencia')
plt.ylabel('Amplitud')
plt.xlim(-10, 10)
plt.grid(True)
plt.show()
```

### 3.2 Obten un resultado Normalizado: 
Investiga de que se trata un resultado normalizado y encargate de normalizar el resultado del punto 3.1

Un resultado normalizado es el proceso donde los valores de un conjunto de datos son escalados de manera que sus características se mantengan constantes o relativas.

En el caso de la transformada de Fourier, normalizar la amplitud de una señal implica ajustar los valores de la transformada para que representen la magnitud de la señal original de manera consistente, independientemente de la escala absoluta de la señal.

```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores de x
x = np.linspace(0, 2 * np.pi, 1000)

# Calcular senal f(x) = sin(x) + cos(2x)
fx = np.sin(x) + np.cos(2 * x)

# Normalizar senal
fx_normalizada = fx / np.max(np.abs(fx))

# Calcular transformada de Fourier
Fourier_x = np.fft.fft(fx_normalizada)

# Obtener frecuencias de la transformada
frec_fourier_x = np.fft.fftfreq(len(x), x[1] - x[0])

# Graficar la transformada de Fourier normalizada
plt.plot(frec_fourier_x, np.abs(Fourier_x), color='black')
plt.title('Transformada de Fourier Normalizada de f(x)')
plt.xlabel('Frecuencia')
plt.ylabel('Amplitud')
plt.xlim(-10, 10)
plt.grid(True)
plt.show()
```

### 3.3 Encuentra la transformada inversa de Fourier
_solo para una señal de las dos que generaste inicialmente_ En este caso puede usar una librería. Explica en detalle que hace la librería y el resultado que arroja.

```python
import numpy as np
import matplotlib.pyplot as plt

# Rango de valores de x
x = np.linspace(0, 2 * np.pi, 1000)

# Calcular señal f(x) = sin(x) + cos(2x)
fx = np.sin(x) + np.cos(2 * x)

# Calcular la transformada de Fourier
Fourier_x = np.fft.fft(fx)

# Transformada inversa de Fourier (ifft)
fx_reconstruida = np.fft.ifft(Fourier_x)        # toma una serie de coeficientes de amplitud y fase de diferentes frecuencias
                                                # y los utiliza para reconstruir la señal original en el dominio del tiempo

# Graficar la señal reconstruida
plt.plot(x, fx_reconstruida.real, color='black', label=r'$ f(x) = \sin(x)+ \cos(2x)$')
plt.title('Transformada Inversa de Fourier de f(x)')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.legend()
plt.show()
```

### 3.4 Escriba un algoritmo para la Transformada Discreta de Fourier(DFT) 2-D y muéstrelo 
_solo para una señal de las dos que generaste inicialmente_
Aquí debe implementar el algoritmo de DFT, es decir, NO puedes usar paquetes externos para calcular la transformada.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define la señal
x = np.linspace(0, 2 * np.pi, 2500)
signal = np.sin(x) + np.cos(2 * x)

# Define la DFT 2D
def DFT2D(signal):
    M, N = signal.shape
    dft2d = np.zeros((M, N), dtype=complex)
    for k in range(M):
        for l in range(N):
            for m in range(M):
                for n in range(N):
                    e = np.exp(- 2j * np.pi * ((k * m) / M + (l * n) / N))
                    dft2d[k, l] += signal[m, n] * e
    return dft2d

# Cambia la forma de la señal a 2D
signal_2d = signal.reshape((50, 50))

# Calcula la DFT 2D
dft2d = DFT2D(signal_2d)

# Grafica la señal original
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(signal_2d, cmap='gray')
plt.title('Señal Original')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

# Grafica el valor absoluto de la DFT
plt.subplot(1, 2, 2)
plt.imshow(np.abs(dft2d), cmap='gray')
plt.title('DFT 2D')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')

plt.show()
```

### 3.5 Explica tus hallazgos
- Explica que diferencia hay entre los dos tipos de ruido generados y el efecto en la onda original.
- Qué representan el resultado de la transformada de fourier para los casos que desarrollaste?

El ruido gaussiano y el ruido rosa difieren principalmente en su distribución espectral. El ruido gaussiano sigue una distribución y densidad espectral plana, lo que significa que todas las frecuencias tienen la misma energía. Por otro lado, el ruido rosa tiene una densidad espectral que disminuye con la frecuencia, lo que significa que las frecuencias más altas tienen menos energía que las frecuencias más bajas. En términos de su efecto en la onda original, el ruido gaussiano puede agregar fluctuaciones aleatorias en todas las frecuencias, mientras que el ruido rosa tiende a afectar más a las frecuencias más altas, suavizando la señal original.

La transformada de Fourier nos proporciona información sobre las frecuencias presentes en una señal. La transformada de Fourier nos muestra cómo las frecuencias se distribuyen en las señales fundamentales y cómo cambian después de la aplicación de operar con ellas, como adición de ruido o la combinación de señales. La magnitud de la transformada de Fourier indica la contribución de cada frecuencia a la señal original, mientras que la fase indica el desfase de cada componente de frecuencia.
